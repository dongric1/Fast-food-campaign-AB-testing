# **1. Introduction & Background**

The fast-food chain is planning to introduce a new item to its menu and aims to determine the most effective marketing strategy to maximize its sales. To achieve this, 3 (three) different campaigns were tested across various markets. Each campaign was assigned to different locations, and weekly sales data for the new item were recorded over the first four weeks of the promotional period. This A/B test evaluates which marketing strategy yields the highest sales and provides actionable insights for the company's marketing decisions.



# **2. Problem  & Goal**

The primary challenge is to select the most effective marketing strategy out of the three tested campaigns to ensure the highest sales for the new menu item. With the potential for substantial impact on revenue, it is crucial to determine which promotion maximizes sales and thus represents the best investment for the company. However when market size and store characteristics differ, treating them as homogeneous could distort test results. Segmenting by market size (small, medium, large) is a must approach to reach robustness.

The objective of this A/B test is to evaluate the performance of the three marketing campaigns and identify the most successful strategy. By analyzing the weekly sales data, the test aims to provide clear recommendations on which promotion should be implemented across all locations to achieve the highest sales increase.

# **3. Dataset Description**

Dataset can be found [HERE](https://www.kaggle.com/datasets/chebotinaa/fast-food-marketing-campaign-ab-test).

Key columns in the dataset are:
* **MarketID**: unique identifier for market.
* **MarketSize**: size of market area by sales.
* **LocationID**: unique identifier for store location.
* **AgeOfStore**: age of store in years.
* **Promotion**: one of three promotions that were tested.
* **Week**: one of four weeks when the promotions were run.
* **SalesInThousands**: sales amount for a specific LocationID, Promotion, and week.

# **4. Target Metric**

Only one metric is availabe in dataset - sales in thousands, which will be named revenue. Revenue is aggregated by location  (see [Appendix 1](#scrollTo=0_Pk6LphMWnZ))

Finally, a two-sided t-test will be used for pairwise comparisons of marketing campaigns. The two-sided t-test is used to determine if there is a significant difference between the means of two groups, without assuming the direction of the difference. This is important when comparing marketing campaigns, as we are interested in detecting any significant difference in either direction (one campaign performs better or worse than another).


# **7. Statistical Tests & Analysis**

Overall, three test groups will be evaluated with the null hypothesis (H₀) assuming that there is no difference between any of the sample groups.

Independent Samples t-test was conducted with a 99 % confifence interval for all groups to avoid multiple testing problem.


image.png

## **Group 1 (1 VS 2)**



```
H1: Promotion 1 generates statistically significantly higher revenue compared to the Promotion 2.
```



image.png


`H1 accepted.`

Promotion 1 (M=232.396, SD = 64.11) generates higher revenue than Promotion 2 (M=189.3177, SD = 57.99).

The two-sided t-test for equality of means indicated significant difference between the two groups, t(88) = 3.347, p = 0.01.

With the mean difference of 43.08 and 99% confidence interval ranging from 9.19 to 76.96, Cohen's d = .706 meaning high effect size (see [Apendix 3](#scrollTo=Dx612gpsViBQ)).



image.png

## **Group 2 (1 VS 3)**


```
H1: Promotion 1 generates statistically significantly higher revenue compared to the Promotion 3.
```



image.png

`H1 rejected. ` There is no statistically significant difference in revenue between Promotion 1 (M = 232.3960, SD = 64.11) and Promotion 3 (M = 221.4579, SD = 65.54).

The two-sided t-test for equality of means indicated no significant difference between the two groups, t(88) = .799, p = .426. With the mean difference of 10.94 and 99% confidence interval ranging from -25.09 to 46.97, which includes zero, suggesting that the true difference in means might not be significantly different.

## **Group 3 (3 VS 2)**

```
H1: Promotion 3 generates statistically significantly higher revenue compared to the Promotion 2.
```


image.png

The two-sided t-test for equality of means indicated no significant difference, t(92) = 2.518, p = .014. The mean difference was 32.14, with a 99% confidence interval ranging from -1.434 to 65.714.

# **8. Results & Findings**



Based on the independent samples t-test results, the following conclusions can be made:

* Promotion 1 vs Promotion 2:

Statistically significant differences exist. Promotion 1 generates significantly higher revenue than Promotion 2.

* Promotion 1 vs Promotion 3:

No statistically significant difference is observed. Promotion 1 does not generate statistically significantly higher revenue compared to Promotion 3.

* Promotion 3 vs Promotion 2:

No statistically significant difference is found at the 99% confidence level. Although Promotion 3 shows higher revenue compared to Promotion 2, the two-sided p-value of 0.014.


In conclusion, it can be said that promotion 1 will generate most revenue compared to promotion 2 (difference 43.08 thousands), however it cannot be trusted that it is the best choice yet as compared with promotion 3 there is no difference, so further tests should be done. However promotion 2 compared to 3 has no difference and compared to promotion 1 performed worse, so bussiness can discontinue promotion 2.


# **9. Suppositions & Limitations**

**Suppositions**

1. **Market Homogeneity**: The analysis assumes that all markets are relatively homogeneous, and the promotional campaigns have a similar impact across different market sizes and store ages.

---
 **Limitations**

1. **Market Differences**: One of the primary limitations is that markets may differ significantly in terms of size, demographics, and other characteristics. The analysis does not account for these potential differences, which could skew the results. To address this limitation, it is advisable to first test for overall differences between promotions. If significant differences are found, further investigations should be conducted by segmenting the data based on market size and other relevant factors such as store age. This approach will help in understanding the nuances of each market and the effectiveness of promotions within different market segments.

2. **Sample sizes** The smaller sample size for Promotion 1 may result in less precise estimates of the mean and increased sensitivity to outliers, which could influence the results compared to Promotions 2 and 3.
